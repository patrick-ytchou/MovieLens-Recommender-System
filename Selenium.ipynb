```python
# Set up result
result = {}

# # close if any
# driver.close()

# Set the web browser
driver = webdriver.Edge(r'driver/msedgedriver', {})
```

```python
i = 0
t0 = datetime.now()
for movie in movies:
    print("Currently scraping movie: ", movie)
    result[movie] = {}
    # Go to imdmb
    driver.get("https://www.imdb.com/")
    time.sleep(1)

    # Search for the movie result
    driver.find_element_by_name("q").send_keys(movie)
    time.sleep(1)
    driver.find_element_by_id("suggestion-search-button").send_keys(Keys.ENTER)
    time.sleep(1)
    driver.find_element_by_xpath('//*[@id="sidebar"]/div[3]/ul/ul/li[1]/a').click()
    time.sleep(1)
    driver.find_element_by_xpath('//*[@id="main"]/div/div[2]/table/tbody/tr[1]/td[2]/a').click()

    # Scrap headline
    try:
        ans = driver.current_url
        page = requests.get(ans)
        soup = BeautifulSoup(page.content, "html.parser")
        headline = soup.find(class_="GenresAndPlot__TextContainerBreakpointXS_TO_M-cum89p-0 dcFkRD").get_text()
        result[movie]['headline'] = headline
    except:
        print("Failed catching headline.")

    # Scrape storyline
    time.sleep(1)
    try:
        storyline = driver.find_element_by_xpath("//div[@data-testid='storyline-plot-summary']").text
#         storyline = driver.find_element_by_xpath('//*[@id="__next"]/main/div/section[1]/div/section/div/div[1]/section[6]/div[2]/div[1]/div[1]/div').text
        result[movie]['storyline'] = storyline 
    except:
        print("Failed catching storyline.") 


    # Scrape movie director
    time.sleep(1)
    try:
        director = driver.find_element_by_xpath('//*[@id="__next"]/main/div/section[1]/div/section/div/div[1]/section[4]/ul/li[1]/div/ul/li/a').text
        result[movie]['director'] = director
    except:
        print("Failed catching director.")

    # Scrape movie Stars
    time.sleep(1)
    try:
        actor_list = []
        for i in range(1,4):
            actor = driver.find_element_by_xpath(f'//*[@id="__next"]/main/div/section[1]/div/section/div/div[1]/section[4]/div[2]/div[{i}]/div[2]/a').text
            actor_list.append(actor)
        result[movie]['actors'] = actor_list
    except:
        print("Failed catching actors.")

    # Scrape top review
    time.sleep(1)
    try:
        review = driver.find_element_by_xpath("//div[@data-testid='review-overflow']").text
#         review = driver.find_element_by_xpath('//*[@id="__next"]/main/div/section[1]/div/section/div/div[1]/section[9]/div[2]/div[1]/div[3]/div[1]/div').text
        result[movie]['top_review'] = review  
    except:
        print("Failed catching review.")


    # get the image source
    try:
        driver.find_element_by_xpath('//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[3]/div[1]/div/div[1]/div/a').click()
        time.sleep(1)
        img = driver.find_element_by_xpath('//*[@id="__next"]/main/div[2]/div[3]/div[4]/img')
        src = img.get_attribute('src')
        urllib.request.urlretrieve(src, f"image/{movie}.png")
    except:
        print("Failed catching image.")
    i += 1
    if i % 15 == 0:
        print(f"Current Status: {i/len(df)*100:.1f}% finished. Time Elapsed: {datetime.now()-t0}.")
        t0 = datetime.now()
```
